```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Define the teams and seasons
teams = [
    "AC", "Alaves", "Almeria","AM", "Barcelona", "Cadiz","CV", "Eibar", "Elche",
    "Espanyol", "Getafe", "Girona", "Granada", "Huesca", "LasPalmas",
    "Levante", "Mallorca", "Osasuna","RB", "RealMadrid", "RealSociadad", "RV",
    "Valencia","Valladolid", "Villarreal"
]
seasons = ["2020-2021", "2021-2022", "2022-2023", "2023-2024"]

# Initialize an empty DataFrame for all players' stats
all_data = pd.DataFrame()

# Load data from each CSV file
for team in teams:
    for season in seasons:
        file_name = f"{team}{season}.csv"
        if os.path.exists(file_name):
            try:
                # Read the CSV file
                df = pd.read_csv(file_name)
                
                # Optional: Add team and season columns for reference
                df['Team'] = team
                df['Season'] = season
                
                # Append to all_data DataFrame
                all_data = pd.concat([all_data, df], ignore_index=True)
            except Exception as e:
                print(f"Error reading {file_name}: {e}")

# Data Cleaning
# Check for missing values
print("Missing values before cleaning:")
print(all_data.isnull().sum())

# Fill or drop missing values (customize as needed)
all_data['Gls'].fillna(0, inplace=True)  # Fill goals with 0
all_data['Ast'].fillna(0, inplace=True)  # Fill assists with 0
all_data.dropna(subset=['Pos', 'Age'], inplace=True)  # Drop rows where Pos or Age is missing

# Convert necessary columns to appropriate types
all_data['Age'] = all_data['Age'].astype(int)
all_data['MP'] = all_data['MP'].astype(int)
all_data['Gls'] = all_data['Gls'].astype(int)
all_data['Ast'] = all_data['Ast'].astype(int)

# Data Analysis - Example: Goals and Assists per Team per Season
summary_stats = all_data.groupby(['Team', 'Season']).agg({'Gls': 'sum', 'Ast': 'sum'}).reset_index()

# Data Visualization
plt.figure(figsize=(12, 8))
sns.barplot(data=summary_stats, x='Team', y='Gls', hue='Season')
plt.title('Total Goals per Team per Season')
plt.xticks(rotation=45)
plt.ylabel('Total Goals')
plt.xlabel('Team')
plt.legend(title='Season')
plt.tight_layout()
plt.show()

# Additional Visualization: Assists per Team per Season
plt.figure(figsize=(12, 8))
sns.barplot(data=summary_stats, x='Team', y='Ast', hue='Season')
plt.title('Total Assists per Team per Season')
plt.xticks(rotation=45)
plt.ylabel('Total Assists')
plt.xlabel('Team')
plt.legend(title='Season')
plt.tight_layout()
plt.show()

# Save cleaned data to a new CSV
all_data.to_csv("cleaned_player_stats.csv", index=False)
print("Cleaned data saved to 'cleaned_player_stats.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Define the teams and seasons
teams = [
    "AC", "Alaves", "Almeria","AM", "Barcelona", "Cadiz","CV", "Eibar", "Elche",
    "Espanyol", "Getafe", "Girona", "Granada", "Huesca", "LasPalmas",
    "Levante", "Mallorca", "Osasuna","RB", "RealMadrid", "RealSociadad", "RV",
    "Valencia","Valladolid", "Villarreal"
]
seasons = ["2020-2021", "2021-2022", "2022-2023", "2023-2024"]

# Initialize an empty DataFrame for all players' stats
all_data = pd.DataFrame()
```

```python
# Load data from each CSV file
for team in teams:
    for season in seasons:
        file_name = f"{team}{season}.csv"
        if os.path.exists(file_name):
            try:
                # Read the CSV file
                df = pd.read_csv(file_name)
                
                # Optional: Add team and season columns for reference
                df['Team'] = team
                df['Season'] = season
                
                # Append to all_data DataFrame
                all_data = pd.concat([all_data, df], ignore_index=True)
            except Exception as e:
                print(f"Error reading {file_name}: {e}")
```

```python
# Data Cleaning
# Check for missing values
print("Missing values before cleaning:")
print(all_data.isnull().sum())

# Fill or drop missing values (customize as needed)
all_data['Performance_Gls'].fillna(0, inplace=True)  # Fill goals with 0
all_data['Performance_Ast'].fillna(0, inplace=True)  # Fill assists with 0
all_data.dropna(subset=['Unnamed: 2_level_0_Pos', 'Unnamed: 3_level_0_Age'], inplace=True)  # Drop rows with missing position or age

# Convert necessary columns to appropriate types
all_data['Unnamed: 3_level_0_Age'] = all_data['Unnamed: 3_level_0_Age'].astype(int)
# all_data['Playing Time_MP'] = all_data['Playing Time_MP'].astype(int)
all_data['Performance_Gls'] = all_data['Performance_Gls'].astype(int)
all_data['Performance_Ast'] = all_data['Performance_Ast'].astype(int)
```

```python
# Data Analysis - Example: Goals and Assists per Team per Season
summary_stats = all_data.groupby(['Team', 'Season']).agg({
    'Performance_Gls': 'sum', 
    'Performance_Ast': 'sum'
}).reset_index()
```

```python
# Data Visualization
plt.figure(figsize=(12, 8))
sns.barplot(data=summary_stats, x='Team', y='Performance_Gls', hue='Season')
plt.title('Total Goals per Team per Season')
plt.xticks(rotation=45)
plt.ylabel('Total Goals')
plt.xlabel('Team')
plt.legend(title='Season')
plt.tight_layout()
plt.show()

# Additional Visualization: Assists per Team per Season
plt.figure(figsize=(12, 8))
sns.barplot(data=summary_stats, x='Team', y='Performance_Ast', hue='Season')
plt.title('Total Assists per Team per Season')
plt.xticks(rotation=45)
plt.ylabel('Total Assists')
plt.xlabel('Team')
plt.legend(title='Season')
plt.tight_layout()
plt.show()
```
